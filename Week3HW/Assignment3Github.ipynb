```python
#I kept the bins as 15 based on the data ...

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mlb


starbucks_set = pd.read_csv("/Users/amoghholla/ashoka-horizons-data-science/Week-3/data/starbucks/starbucks.csv")

plt.figure(figsize=(8, 5))
plt.hist(starbucks_set['Calories'], bins=15, color='lightyellow', edgecolor='black')
plt.title("Calories in Starbucks drinks")
plt.xlabel("Calories")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8, 3))

sns.boxplot(x=starbucks_set['Calories'], color='lightgreen')
plt.title("Box Plot of Calories in Starbucks Drinks")
plt.xlabel("Calories")

plt.show()
```

The interpretation is that: 

The historgram gives information on the frequency distribution on calories showing the highest calorie containing products. 

The box plot on the other hand shows the outliers, the interquartile range, the median calorie value in the dataset and the highest and lowest calorie values. 


```python

sns.countplot(data=netflix_set, x='type', palette='Set1')


plt.title('Frequency of Netflix Content ', fontsize=20)
plt.xlabel('Content Type', fontsize = 10)
plt.ylabel('Count', fontsize = 10)

plt.grid(axis='y', linestyle='--')


plt.show()
```

```python
sns.kdeplot(data = diabetes_set, x = 'Age', label='KDE of Age')


sns.histplot(data = diabetes_set, x = 'Age', kde = True, color = 'red', )


plt.title('Distribution of Age in PIMA Diabetes Dataset')
plt.xlabel('Age')
plt.ylabel('Density')


plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=starbucks_set, x='Calories', y=' Sugars (g)')


plt.title('Relationship between Calories and Sugar in Starbucks Menu Items')
plt.xlabel('Calories')
plt.ylabel('Sugar (g)')

plt.show()

# Interpretation of the relationship

print("Interpretation: This scatter plot visually represents the relationship between the calorie content and sugar content (in grams) of various Starbucks menu items.  The spread of the data points will indicate how strong this relationship is across different items.")
```

```python
sns.boxplot(data=diabetes_set, x='Outcome', y='Glucose')


plt.title('Distribution of Glucose Levels by Diabetes Outcome')
plt.xlabel('Outcome (0: No Diabetes, 1: Diabetes)')
plt.ylabel('Glucose Level')
plt.xticks([0, 1], ['No Diabetes', 'Diabetes']) 

plt.show()

# Interpretation of the relationship
print("Interpretation: This box plot suggests a clear difference in glucose levels between individuals with and without diabetes. We can observe that individuals with an 'Outcome' of 1 (Diabetes) tend to have significantly higher median glucose levels compared to those with an 'Outcome' of 0 (No Diabetes). ")
```

```python
#I just made titles per year as tpy

tpy = netflix_set['release_year'].value_counts().sort_index().reset_index()
tpy.columns = ['Release Year', 'Number of Titles']


sns.lineplot(data = tpy, x='Release Year', y='Number of Titles')

plt.title('Total Number of Netflix Titles Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')

plt.show()

#I learnt some of the functions from chatGPT

movie_tv_counts = netflix_set.groupby(['release_year', 'type']).size().unstack(fill_value=0).reset_index()
movie_tv_counts.columns.name = None 

sns.lineplot(data=movie_tv_counts, x='release_year', y='Movie', label='Movies')
sns.lineplot(data=movie_tv_counts, x='release_year', y='TV Show', label='TV Shows')


plt.title('Number of Netflix Movies and TV Shows Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend(title='Content Type')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.scatterplot(data=diabetes_set,x='BMI',y='Age',hue='Outcome',size='Pregnancies',sizes=(20, 400),)
plt.title('BMI vs. Age, Colored by Outcome and Sized by Pregnancies')
plt.xlabel('BMI')
plt.ylabel('Age')
plt.legend(title='Outcome (0: No Diabetes, 1: Diabetes)')

plt.show()
```

```python
selected_features = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
correlation_matrix = diabetes_set[selected_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix,
    annot = True,
    cmap = 'coolwarm',
    fmt = ".2f",
    linewidths = 5
)
plt.title('Correlation Heatmap of Selected PIMA Dataset Features')
plt.show()
```

The interpretation: 

The highly correlated features include Age and Preganacies with the highest correlation of 0.54... So this means as the age increases the preganancies will also increases. 

The lowly correlated features are several for instance skint thickness and age, where it has a negative correlation of -0.11. This shows that there is no related at all in terms of skin thickness and age. 

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))


sns.histplot(data=diabetes_set, x='BloodPressure', kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Blood Pressure (PIMA)')
axes[0, 0].set_xlabel('Blood Pressure')
axes[0, 0].set_ylabel('Frequency')


sns.scatterplot(data=starbucks_set, x='Calories', y=' Protein (g) ', ax=axes[0, 1])
axes[0, 1].set_title('Calories vs. Protein (Starbucks)')
axes[0, 1].set_xlabel('Calories')
axes[0, 1].set_ylabel('Protein (g)')


sns.countplot(data=starbucks_set, x='Beverage_category', ax=axes[1, 0])
axes[1, 0].set_title('Count of Starbucks Beverage Categories')
axes[1, 0].set_xlabel('Beverage Category')
axes[1, 0].set_ylabel('Count')
axes[1, 0].tick_params(axis='x', rotation=45) 


sns.boxplot(data=diabetes_set, x='Outcome', y='Age', ax=axes[1, 1])
axes[1, 1].set_title('Age Distribution by Diabetes Outcome (PIMA)')
axes[1, 1].set_xlabel('Outcome (0: No Diabetes, 1: Diabetes)')
axes[1, 1].set_ylabel('Age')
axes[1, 1].set_xticklabels(['No Diabetes', 'Diabetes'])



fig.suptitle('Multivariate Analysis in Starbucks and Diabetes datasets', fontsize=16)
plt.tight_layout()
plt.show()
```
